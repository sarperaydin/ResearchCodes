```python
class Agents:
    
    def __init__(self,N,ps,state,network,gamma,w_self):
            #number of agents
            self.N=N
            #policy parameter size
            self.ps=ps
            #random policy initilization
            self.policy=-(0.001)*np.random.random((N,ps))
            #self.policy=-rng.rand((N,ps))
            #local info of i
            self.local=np.repeat(self.policy[:, :, np.newaxis], N, axis=2)
            #state
            #self.state=np.atleast_1d(state)
            self.state=state
            #total rewards
            #self.total_rewards=np.zeros((N,tf))
            #rewards inside episode
            self.reward_eps=np.zeros(N)
            self.actions=np.zeros(N)
            #communication network
            self.network=network
            self.w_all=self.generate_weights(w_self)
            #discount rate
            self.gamma=gamma
            
            ## local difference stats
            
            self.local_dif=np.zeros(N)
            
            
        #Episode    
    def eps_run(self,s_0,T_1,T_2,b,q,gamma):
        
            #State Initilization
            self.state=s_0
            a=self.sample_actions(self.state)

            #First episode to compute gradient of log-policy
            for t in range(T_1):
                #State Transition-State Dynamics
                self.state=self.state_dy(self.state,a,b,q)
                #Action Sampling 
                a=self.sample_actions(self.state)

            
            a_hist=np.copy(a)
            #print('print: '+str(a_hist))
            s_hist=np.copy(self.state)
            Q=np.zeros(self.N)
        
            #Reward Collection
            for t in range (T_2):

                r=self.reward_func(a,self.state)
                Q=Q+((gamma)**(t/2))*r
                self.state=self.state_dy(self.state,a,b,q)
                a=self.sample_actions(self.state)
               
            
            r=self.reward_func(a,self.state)
            Q=Q+((gamma)**(t/2))*r
        
            output = {'Q':Q, 'a':a_hist,'s':s_hist}
            
            return output
    
    # Linear combination between parameters and states        
    def lin_mean(self,s):
        
        #aug_s=np.zeros(self.ps)
        means=np.zeros(self.N)
        for i in range(self.N):
            
            local_i=self.local[:,:,i]
            local_ij=np.delete(local_i,i,0)
            sum_j=np.sum(local_ij[:,0])
            
            #aug_s=np.array([s,s*sum_j])
            aug_s=np.array([s,sum_j])
            means[i]=np.dot(aug_s,self.policy[i,:])
            
        return means
            
            
    #Action is sampled with clipped policies.
    def sample_actions(self,s):
        
        low=0.1
        upp=10
        sd=0.1
        
        actions=np.zeros(self.N)
        lin_means=self.lin_mean(s)
        act_mean= 1/(1 + np.exp(-lin_means))
        cov=sd*np.identity(self.N)
        actions=np.random.multivariate_normal(act_mean, cov)
        actions[actions>1]=1
        actions[actions<0]=0
            #self.actions[i,:]
        #print(actions) 
        return actions
    
    #State Dynamics                
    def state_dy(self,s,a,b,q):
        
        

        prob=np.random.binomial(1, 1, 1)
        prob=np.asscalar(prob)
        #print(np.sum(a))
        state= b*s+(s**q/(1+s**q))+prob*np.sum(a)
        #state=min(state,10)
        #print(state)
        return state
        
    #Reward Function and they are bounded satisfy the assumption.
    def reward_func(self,a,s):

        rewards=np.log10((10**4)*a+1)-s**2

        rewards[rewards<-4]=-4
        rewards[rewards>4]=4

        return rewards
    
    #Sigmoid maps to actions space in [0,1]
    def sigmoid(self,x):
    
        s= 1/(1 + np.exp(-x))
        return s
    
    # Gradient of Normal Distribution
    def grad_normal(self,a,mean,sd):
        
        grad_n=0
        
        if a>0 and a<1:
        
            grad_n=(mean-a)/sd
            
        elif a==0:
            
            a_std=(a-mean)/sd
            grad_n=(-1/sd)*multivariate_normal.pdf(a_std)
            grad_n=grad_n/multivariate_normal.cdf(a_std)
            
        elif a==1:
            
            a_std=(a-mean)/sd
            grad_n=(-1/sd)*multivariate_normal.pdf(a_std)
            
            div=(1-multivariate_normal.cdf(a_std))
            
            if(div < 10**(-5)):
                grad_n=-grad_n/10**(-5)
            else:
                 grad_n=-grad_n/div
        
        return grad_n
    
    # Gradient of Sigmoid
    def grad_sig(self,x):
        
        sig=self.sigmoid(x)
        gs=sig*(1-sig)
        
        return gs
    
    # Gradient of the Linear part
    def grad_m(self,i,j,s):
        
        grad=0
        
        if (i==j):
            
            masked_i=np.ma.array(self.local[:,1,i], mask=False)
            masked_i.mask[i]=True
            sum_j2=masked_i.sum()
            grad=np.array([s, (1/(self.N-1))*sum_j2])
        
        
        else:
            theta_ji=self.local[j,1,i]
            grad=np.array([(1/(self.N-1))*theta_ji,0])
        
        return grad
    
    
    #Gradient Update
    def update_grad(self,output,alpha):
        
        np.seterr(divide='raise')
        
        grad=np.zeros((N,self.ps))
        s=output['s']
        a=output['a']
        Q=output['Q']
        

        norm_g=np.zeros(self.N)
        sd=0.1
        lin_means=self.lin_mean(s)
        mean_act=1/(1 + np.exp(-lin_means))

        for i in range(self.N):
            
            grad_i=np.array([0,0])
            for j in range(self.N):
                
                p1=self.grad_normal(a[j],mean_act[j],sd)
                p2=self.grad_sig(lin_means[j])
                p3=self.grad_m(i,j,s)
                grad_i=grad_i+p1*p2*p3

            self.policy[i]=self.policy[i]+alpha*(1/(1-self.gamma))*Q[i]*grad_i
            self.local[i,:,i]=self.policy[i]
            norm_g[i]=(1/(1-self.gamma))*abs(Q[i])*LA.norm(grad_i)
            
        #print(norm_g)
        return norm_g
    
    # Multi-Agent and Networked Polciy Gradient Play
    def multi_PG(self,T_max):
        
        #State Dynamics and initial state definitions
        s_0=0
        b=0.4
        q=2

        
        grad_norm=np.zeros((self.N,T_max))
        
        #Stats
        self.local_dif=np.zeros(T_max)
        self.Q=np.zeros((N,T_max))
        #Random horizon generation
        eps_lens=np.random.geometric(1-self.gamma**(0.5), (2,T_max))
        for t in range(T_max):
            
            T_1=eps_lens[0,t]
            T_2=eps_lens[1,t]
            #Episodic Learning
            output=self.eps_run(s_0,T_1,T_2,b,q,self.gamma)
            self.Q[:,t]=output['Q']
            
            #Stepsize
            alpha=0.001/(t+1)
            #Gradient Update
            grad_norm[:,t]=self.update_grad(output,alpha)
            self.share_local()
            
            self.local_dif[t]=self.getLocalDif()
        
 
            
        if (np.isnan(np.min(grad_norm))): 
            print('There is a NaN')
        else:
            self.plotGrad(grad_norm,T_max)
            self.plot_local(T_max,self.local_dif)
        return grad_norm 

    #Local Error at each time step computed with this functions.
    def getLocalDif(self):
        
        
        total_dif=0
        for i in range(self.N):
            
  
            others_i=self.local[i,:,:].T
            dif_i=self.policy[i]-others_i
            norm_i=LA.norm(dif_i, axis=1)
            sum_i=norm_i.sum()
            
            total_dif=total_dif+sum_i
            
        total_dif=(1/(self.N*(self.N-1)))*total_dif
        
        return total_dif
    
    
    #Information sharing among agents via consensus and row stochactic matrices.   
    def share_local(self):
        
        for i in range(self.N):
            
            for j in range(self.N):
                self.local[j,:,i]=np.matmul(self.w_all[j,:,i],self.local[j,:,:].T)
            
            
            
        
    # Weight generation for communication    
    def generate_weights(self,w_self):
        
        w_all=np.zeros((self.N,self.N,self.N))
        ## In self. network we asssume self_network[i,i]=0
        for i in range(self.N):
            
            com_i=self.network[i,:]
            com_yes=com_i==1
            com_false=com_i==0
            
            noa=com_yes.sum()
            no_false=com_false.sum()
            
            
            w_all[i,i,i]=1
            w_all[com_yes,com_yes,i]=1
            
            if(no_false>0):
                w_all[com_false,i,i]=w_self
            if (no_false>0) and (noa>0):
                w_all[com_false,com_yes,i]=(1-w_self)/noa
            
            
        return w_all
        
    #Gradient Plotting    
    def plotGrad(self,grad_norm,T_max):
        
        
        fig = plt.figure()
        ax = fig.gca()
        ax.set_xticks(np.arange(0, 2*T_max, T_max/5))
        avg_norm=np.mean(grad_norm,axis=0)
        ax.set_yticks(np.arange(0, ceil(np.max(avg_norm)),ceil(np.max(avg_norm))/5))
        #plt.grid(linestyle='--')
        plt.grid(b=True, which='major', color='k', linestyle='-',alpha=0.8)
        plt.grid(b=True, which='minor', color='silver', linestyle='--', alpha=0.2)
        plt.minorticks_on()
        #plt.style.use('fivethirtyeight')
        plt.style.use('default')
        font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }
        #font.set_style('italic')
        ax.set_yscale('log')
        #plt.yscale("log") 
        #col_list=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
        col_list=['#0051a2', '#97964a', '#ffd44f', '#f4777f', '#93003a']
        #for i in range(self.N):
        #agent_i="Agent "+str(i+1)
            
        plt.plot(range(1,T_max+1),avg_norm,label='Average Norm of Gradient',color=col_list[0])
            #plt.plot(range(1,T_max+1),grad_norm[1,:],label="Agent 2",color='#0000CC')
            #plt.plot(range(1,T_max+1),grad_norm[2,:],label="Agent 3",color='#33cc33')
    
        plt.legend(loc='upper right',fontsize=14)
        
        plt.gca().spines['top'].set_visible(False)
        plt.gca().spines['right'].set_visible(False)
        #plt.ticklabel_format(axis='both', style='sci', scilimits=(0,0))
        #fig.suptitle('test title', fontsize=20)
        plt.xlabel('Iteration', fontsize=16,fontdict=font)
        plt.ylabel('Norm of Gradient', fontsize=16,fontdict=font)
        plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0e}')) # 2 decimal places
        #plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:, .2.2e}')) 
        
        
        plt.xlim([0, T_max+1])
        #plt.ylim([0, np.max(grad_norm)])

        plt.show() 
        

    #Local Error Plotting
    def plot_local(self,T_max,local_dif):
        
        
        fig = plt.figure()
        ax = fig.gca()
        ax.set_xticks(np.arange(0, 2*T_max, T_max/5))
        #avg_norm=np.mean(grad_norm,axis=0)
        avg_norm=local_dif
        ax.set_yticks(np.arange(0, ceil(np.max(avg_norm)),ceil(np.max(avg_norm))/5))
        #plt.grid(linestyle='--')
        plt.grid(b=True, which='major', color='k', linestyle='-',alpha=0.8)
        plt.grid(b=True, which='minor', color='silver', linestyle='--', alpha=0.2)
        plt.minorticks_on()
        #plt.style.use('fivethirtyeight')
        plt.style.use('default')
        font = {'family': 'serif',
                'color':  'black',
                'weight': 'normal',
                'size': 16,
                          }
        #font.set_style('italic')
        ax.set_yscale('log')
        #plt.yscale("log") 
        #col_list=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
        col_list=['#0051a2', '#97964a', '#ffd44f', '#f4777f', '#93003a']
        #for i in range(self.N):
        #agent_i="Agent "+str(i+1)
            
        plt.plot(range(1,T_max+1),avg_norm,label='Ring Network',color=col_list[0])
        #plt.plot(range(1,T_max+1),grad_norm[1,:],label="Agent 2",color='#0000CC')
        #plt.plot(range(1,T_max+1),grad_norm[2,:],label="Agent 3",color='#33cc33')
    
        plt.legend(loc='upper right',fontsize=14)
        
        plt.gca().spines['top'].set_visible(False)
        plt.gca().spines['right'].set_visible(False)
        #plt.ticklabel_format(axis='both', style='sci', scilimits=(0,0))
        #fig.suptitle('test title', fontsize=20)
        plt.xlabel('Iteration', fontsize=16,fontdict=font)
        plt.ylabel('Average Estimation Error', fontsize=16,fontdict=font)
        plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0e}')) # 2 decimal places
        #plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:, .2.2e}')) 
        
        
        plt.xlim([0, T_max+1])
        #plt.ylim([0, np.max(grad_norm)])

        plt.show() 
        
        
        
        
            
```

```python
#Ring Communication Network
def ringNet(N):
    network=np.zeros((N,N))
    np.fill_diagonal(network, -1)
    for i in range(N):
        if(i<N-1):
            network[i,i+1]=1
        else:
            network[N-1,0]=1

    return network
```

```python
#Main Method
N=5 #Number of agents
ps=2 # Number of policy parameters for each agent
state=0 #Intial State
gamma=0.99 # Discount
w_self=0.30 # The weight each agents assigsn on itself for information sharing
network= ringNet(N) #ring network Initilization
T_max=20 #Number of iterations
norm=np.zeros((N,T_max)) # Storage for norm of gradient

#Number of replicatons
rep=1

local_dif=np.zeros(T_max)
num_nan=0
for k in range(rep):
    ags=Agents(N,ps,state,network,gamma,w_self)
    norm_each=ags.multi_PG(T_max)
    #A
    avg_norm=np.mean(norm_each,axis=0)

    dif_each=ags.local_dif
    if(np.isnan(np.min(norm_each))):
        num_nan=num_nan+1
    else:
        norm=norm+norm_each
        local_dif=local_dif+dif_each
    

print('Final Average')
print('Number of NaN: '+str(num_nan))
ags.plotGrad((1/rep)*norm,T_max)
ags.plot_local(T_max,(1/rep)*local_dif)
```
